### Split and explore both Categorical and Numerical data

```python
numerical_cols=[col for col in data.columns if data[col].dtype!=object]
categorical_cols=[col for col in data.columns if data[col].dtype==object]
```

```python
categ_feats = df.select_dtypes('object')
print("No of Categorical features =",categ_feats.shape[1])
```

```python
numeric_feats = df.select_dtypes(exclude =['object'])
print("No of Numerical features =",numeric_feats.shape[1])
```

## Exploring Numerical data

```python
# A function to explore the numerical data in terms of NaNs, Skewness, Histogram plot for each column
def Numerical_Summary():
    for col in numerical_cols:
        print(col)
        print("No of Missing Values =",data[col].isnull().sum())
        print("Skew =",data[col].skew())
        # print("Histogram")
        # fig=px.histogram(data_frame=data,x=col)
        # fig.show()
        print("-"*60)
```

```python
Numerical_Summary()
```

## Exploring Categorical data

```python
# A function to explore the categorical data in terms of NaNs, Value counts , a pie chart for each column
def Categorical_Summary():
    for col in categorical_cols:
        print(col)
        print("No of Missin Values =",data[col].isnull().sum())
        print(data[col].value_counts())
        print("-"*60)
        
```

```python
Categorical_Summary()
```

```python
categ_feats
```

```python
np.array(categ_feats.value_counts())[0]
```

```python
# categ_feats.plot.pie(figsize=(25, 15), subplots = True)
# categ_feats.value_counts().plot(kind='bar', subplots=True, fontsize=17, layout=(2,2), figsize=(10,10))
for col in categ_feats.columns:
  print(col)
  categ_feats[col].value_counts().plot(kind='bar', subplots=True, fontsize=17, figsize=(7,5))
  plt.show()
  print('-'*60)
  plt.close()
```

### Insights about numerical data:
1. Male lenders are way more than females
2. Singles are more likely to take a loan than married
3. Fully employed and with more than 5 years are the dominant
4. Most of lenders are properties owners
5. Most popular language is Estonian
